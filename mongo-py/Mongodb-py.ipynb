```python
from pymongo import MongoClient

# Requires the PyMongo package.
# https://api.mongodb.com/python/current

client = MongoClient('mongodb://localhost:27017/')
filter={
    'names.imdb_name_id': 'nm0000375', 
    'year': {
        '$gt': 2000
    }
}
project={
    '_id': 0
}
sort=list({
    'avg_vote': -1
}.items())

result = client['movie']['movies'].find(
  filter=filter,
  projection=project,
  sort=sort
)
```

```python
import pandas as pd
df = pd.DataFrame(list(result))
df
```

```python
#พล็อตกราฟ
#df.set_index('title')['avg_vote'] #กำหนดเเกน x,y
df.set_index('title')['avg_vote'].plot.bar(figsize=(10,5))
```
